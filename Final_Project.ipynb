```python
from IPython import get_ipython
from IPython.display import display
# %%
!pip install fpdf
from fpdf import FPDF
import pandas as pd
from IPython.display import display, FileLink
from google.colab import files
import os
# %%
input_type = input("Enter the type of input ('MCQ' or 'Code'): ").strip().lower()
# %%
if input_type == "mcq":
    input_mode = input("Choose input method: 'upload' or 'direct': ").strip().lower()

    # Ask for custom Excel column titles
    print("Enter the column titles separated by commas (e.g., Question,Option A,Option B,Option C,Option D,Answer)")
    column_titles_input = input("Titles: ")
    column_titles = [t.strip() for t in column_titles_input.split(',')]

    # Gather MCQ data
    if input_mode == "upload":
        uploaded = files.upload()
        file_name = list(uploaded.keys())[0]
        # It seems you were trying to read a text file in upload mode.
        # If the user uploads an Excel file, this will cause an error later.
        # You'll need to adjust this based on whether you expect a text file or an Excel file upload.
        # For now, assuming a text file where each line is a question.
        try:
            with open(file_name, 'r') as file:
                mcq_data = file.readlines()
        except Exception as e:
            print(f"Error reading uploaded file: {e}")
            mcq_data = [] # Initialize as empty to prevent further errors
    else:
        # Added error handling for the number of questions input
        while True:
            try:
                n_str = input("How many questions? ")
                n = int(n_str)
                if n <= 0:
                    print("Please enter a positive number.")
                    continue # Ask again if the number is not positive
                break # Exit loop if input is a valid positive integer
            except ValueError:
                print("Invalid input. Please enter an integer number.")

        if n > 0:
            print(f"Enter each question in the format: {' | '.join(column_titles)}")
            mcq_data = [input(f"Q{i+1}: ") for i in range(n)]
        else:
            mcq_data = [] # No questions to input

    # Parse MCQs into DataFrame
    parsed_data = []
    for line in mcq_data:
        parts = [p.strip() for p in line.strip().split('|')]
        if len(parts) >= len(column_titles):
            parsed_data.append(parts[:len(column_titles)])
        else:
             print(f"Skipping line due to insufficient columns: {line.strip()}")


    # Check if parsed_data is empty before creating DataFrame
    if not parsed_data:
        print("No valid data to create DataFrame. Skipping file save.")
    else:
        df = pd.DataFrame(parsed_data, columns=column_titles)

        # Save to .xls file
        # Note: The 'xls' engine is deprecated and might require specific libraries (like xlwt).
        # Consider using 'xlsx' engine which is more common.
        output_file = "mcq_output.xlsx" # Changed to xlsx for modern compatibility
        try:
            # You might need to install openpyxl: !pip install openpyxl
            df.to_excel(output_file, index=False, engine='openpyxl')
             # Download the file
            print("✅ MCQs saved as .xlsx file:")
            files.download(output_file)
        except Exception as e:
             print(f"Error saving or downloading file: {e}")
elif input_type == "code":
    input_mode = input("Choose input method: 'upload' or 'direct': ").strip().lower()

    if input_mode == "upload":
        uploaded = files.upload()
        file_name = list(uploaded.keys())[0]
        with open(file_name, 'r') as file:
            code = file.read()
    else:
        print("Enter your code (end with a blank line):")
        lines = []
        while True:
            line = input()
            if line.strip() == "":
                break
            lines.append(line)
        code = "\n".join(lines)

    # Save code to PDF
    class PDF(FPDF):
        def __init__(self):
            super().__init__()
            self.set_auto_page_break(auto=True, margin=15)
            self.add_page()
            self.set_font("Courier", size=12)

        def add_code(self, text):
            for line in text.split('\n'):
                self.cell(0, 10, txt=line, ln=True)

    pdf = PDF()
    pdf.add_code(code)
    output_file = "code_output.pdf"
    pdf.output(output_file)

    print("✅ Code saved as PDF:")
    files.download(output_file)

else:
    print("❌ Invalid input type. Please choose 'MCQ' or 'Code'.")
```
